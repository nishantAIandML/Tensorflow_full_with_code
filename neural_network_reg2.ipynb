```python
import tensorflow as tf
```

```python
synthetic_data=[]
synthetic_data_ans=[]
for i in range(100):
    synthetic_data.append([i,i+1])
    synthetic_data_ans.append([2*i+1])
# synthetic_data
synthetic_data=tf.constant(synthetic_data)
synthetic_data_ans=tf.constant(synthetic_data_ans)
```

```python
#  create a model
model=tf.keras.Sequential([
    tf.keras.layers.Dense(100,activation='relu',input_shape=(2,)),
    tf.keras.layers.Dense(200,activation='relu'),
    # tf.keras.layers.Dense(200,activation='relu'),
    tf.keras.layers.Dense(100,activation='relu'),
    tf.keras.layers.Dense(1)
])

#  compiling model
model.compile(
    loss=tf.keras.losses.mae,
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
    metrics=['mse']
)

# fitting dataset into model
model.fit(synthetic_data,synthetic_data_ans,epochs=100)
```

```python
#  learning rate is most import as compared to other parameters for tuning
sample_tensor=tf.constant([[3,4],[4,9]])
model.predict(sample_tensor)
```

```python
#  Evaluating a model
#  The 3 sets
#  Training(70-80%)
#  Validation(10-15%)
#  Test set(10-15%)
model.summary()
```
